```python
import numpy as np
import noise
import matplotlib.pyplot as plt
import random
```

```python

# word dim
X = 250
Y = 250


pn = np.zeros((X, Y))

# zoom
scale = 100.0

# layers of detail
octaves = 25
# effects of each layer (should be < 1)
persistence = 0.3
# increasing frequency of each layer (should be > 1)
lacunarity = 5.0

seed = random.randint(0, 50)

for x in range(X):
    for y in range(Y):
        pn[x,y] = noise.pnoise2(x/scale, y/scale, octaves=octaves, persistence=persistence, lacunarity=lacunarity, base = seed)

plt.figure(figsize=(10,10))
plt.imshow(pn, cmap="terrain")
plt.show()
```
